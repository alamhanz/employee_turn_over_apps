# Employee Turn Over - Data Cleansing

Cleaning the data interim we have

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
import joblib
warnings.filterwarnings("ignore")
import os
```

```python
PATH_DATA='../../data/raw/'
PATH_DATA_INTERIM='../../data/iter2/interim/'
PATH_MODEL = '../../model/iter2/'
```

## Get Data

```python
df1 = pd.read_csv(PATH_DATA_INTERIM+'iter2_interim.csv')
```

```python
df1.sample(10)
```

```python
df1.shape
```

## Imputation

```python
col_null = df1.isnull().sum()/len(df1)
col_null[col_null>0]
```

```python
df2 = df1.fillna(0)
```

```python
df2
```

## Outliers

```python
col_out = set(df2.columns)-set(['unique_employee_ids','stillExists','companyAlias','is_train'])
df2_train = df2[df2.is_train == 1]
```

```python
outlier_rules = df2_train[col_out].quantile(0.98).to_dict()
joblib.dump(outlier_rules, PATH_MODEL+'outliers_rules.pkl')
```

```python
outlier_rules
```

```python
df3 = df2.copy()
for c in col_out:
    df3[c] = np.where(df3[c]>=outlier_rules[c], outlier_rules[c], df3[c])
```

```python
df3
```

```python
df3.to_csv(PATH_DATA_INTERIM+'iter2_interim_cln.csv',index = False)
```
